# Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Read the synthetic data

```python
data1 = pd.read_excel("Sintetik_Univariate Linear.xlsx")
data1
```

# Spliting Data

```python
X = np.array(data1.X)

Y = np.array(data1.Y)
```

```python
X.shape
```

```python
plt.plot(X,Y,"ro")
```

# Learning Algorithm (Initialize + Gradient Descent)

```python
#Initialize a zero value
w1 = 0
w0 = 0
h = np.dot(X,w1) + w0
err = ((h-Y))
J = (err**2).mean()/2
J0 = []
iterations = 10**5

for i in range(iterations):
    dw1 = (err*X).mean()
    dw0 = (err).mean()
    
    w1 = w1 - 0.05*dw1
    w0 = w0 - 0.05*dw0
    
    h = X*w1 + w0
    
    err = ((h-Y))
    
    J = (err**2).mean()/2
    
    if J<0.001 :
        break
            
    else :
        if i&1000==0:
            print(J)
            J0.append(J)
           
  
```

```python
print(w1,w0)
```

```python
plt.plot(J0)
plt.title("Learning each Iteration")
```

```python
plt.plot(X,Y,"ro")
plt.plot(X,h.T)
```

# Ordinary Least Squared

```python
x = X.reshape(75,1)
y = Y.reshape(75,1)
a = np.zeros((75,1))+1
```

```python
x =  np.c_[x,a]
```

```python
w = np.dot(np.linalg.inv(np.dot(x.T,x)),np.dot(x.T,y))
w
```

```python
h2 = np.dot(x,w)

plt.plot(X,Y,"ro")
plt.plot(X,h2,"k")
plt.plot(X,h,"g")
plt.legend(["Data","OLS","GD"])
```
